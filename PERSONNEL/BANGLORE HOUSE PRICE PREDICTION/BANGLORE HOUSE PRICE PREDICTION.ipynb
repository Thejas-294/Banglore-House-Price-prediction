```python
banglore_house_price_df_new = banglore_house_price_df.drop(['area_type','society','balcony','availability'],axis = 'columns')
banglore_house_price_df_new.head(25)
```

```python
banglore_house_price_df_new.isnull().sum()
```

```python
bath_col_median = banglore_house_price_df_new[['bath']].median()
banglore_house_price_df_new['bath'].fillna(bath_col_median,inplace = True)
banglore_house_price_df_new['size'].fillna(0,inplace = True)
```
